```python
# BY AGE

SELECT AGE_category, 
    COUNT(AGE_category), 
    COUNT(AGE_category)*100/(SELECT COUNT(*) FROM New 
    WHERE PROD_TYPE == 'THRIVE') AS Percentage
FROM(
    SELECT *, 
        CASE WHEN AGE <30 AND AGE >= 0 THEN 'AGE < 30'
        WHEN AGE <60 AND AGE >= 30 THEN '30 < AGE < 60'
        ELSE 'AGE >= 60' END AS 'AGE_category'
    FROM New)
WHERE PROD_TYPE != 'THRIVE'
GROUP BY AGE_category

```

```python
# BY Balance
SELECT Balance_category, 
    COUNT(Balance_category), 
    COUNT(Balance_category)*100/(SELECT COUNT(*) FROM New) AS Percentage
FROM(
    SELECT *, 
    CASE WHEN BALANCE < (SELECT MAX(BALANCE)/4 FROM new) AND BALANCE >= (SELECT MIN(BALANCE) FROM new) THEN 'Balance 1'
  
    WHEN BALANCE <(SELECT MAX(BALANCE)*2/4 FROM new) AND BALANCE >= (SELECT MAX(BALANCE)/4 FROM new) THEN 'Balance 2'
  
    WHEN BALANCE <(SELECT MAX(BALANCE)*3/4 FROM new) AND BALANCE >= (SELECT MAX(BALANCE)*2/4 FROM new) THEN 'Balance 3'
  
    ELSE 'Balance 4' END AS 'Balance_category'
    FROM New)

WHERE PROD_TYPE != 'THRIVE'
GROUP BY Balance_category


```

```python
%%sql

CREATE TABLE tenure AS 
SELECT year_of_joining, 
    COUNT(year_of_joining) AS count
FROM(
    SELECT (SELECT max(STRFTIME('%Y', CLIENT_DATE)) FROM df) - STRFTIME('%Y', CLIENT_DATE) AS year_of_joining
    FROM df) AS sub
GROUP BY year_of_joining
```

```python
%%sql

SELECT *, (count*100/(SELECT sum(count) FROM tenure)) AS percentage FROM tenure
```

```python
CREATE TABLE new1 AS 
SELECT *, 
    CASE WHEN NUM_OF_CONTACT <10 AND NUM_OF_CONTACT >= 0 THEN 'NUM_OF_CONTACT < 10'
    WHEN NUM_OF_CONTACT <20 AND NUM_OF_CONTACT >= 10 THEN '10 < NUM_OF_CONTACT < 20'
    ELSE 'NUM_OF_CONTACT >= 20' END AS 'CONT_category'
FROM new


SELECT PROD_TYPE, 
    CONT_category, 
    COUNT(CONT_category) 
FROM new1
GROUP BY PROD_TYPE, CONT_category
ORDER BY COUNT(CONT_category) DESC


SELECT PROD_TYPE, 
    CONT_category, 
    COUNT(CONT_category)
FROM new1
WHERE PROD_TYPE != 'THRIVE'
GROUP BY PROD_TYPE, CONT_category
ORDER BY COUNT(CONT_category) DESC


SELECT PROD_TYPE, 
    EmailPermission, 
    COUNT(EmailPermission) 
FROM new1
WHERE PROD_TYPE != 'THRIVE'
GROUP BY PROD_TYPE, EmailPermission
ORDER BY COUNT(EmailPermission) DESC


SELECT PROD_TYPE, 
    MailPermission, 
    COUNT(MailPermission)  
FROM new1
WHERE PROD_TYPE != 'THRIVE'
GROUP BY PROD_TYPE, MailPermission	
ORDER BY COUNT(MailPermission) DESC
```
